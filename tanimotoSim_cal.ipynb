```python
#sample running
import pandas as pd
from rdkit import Chem
from rdkit.Chem import AllChem
from rdkit.Chem import DataStructs

# 데이터 프레임 읽기
data_ref = pd.read_csv('/home/soowon/project/07_PPAR/ref_setA.csv')
data_enam = pd.read_csv('/home/soowon/Database/Enamine/SCR_2310/screen_lib2310.csv')

# 결과를 저장할 리스트
results = []

# data_ref의 모든 SMILES에 대해 반복
for index, ref_row in data_ref.iterrows():
    ref_smiles = ref_row['smiles']
    ref_mol = Chem.MolFromSmiles(ref_smiles)
    ref_fp = AllChem.GetMorganFingerprintAsBitVect(ref_mol, 2)

    # data_enam의 모든 SMILES에 대해 반복
    for enam_index, enam_row in sampled_data.iterrows():
        enam_smiles = enam_row['SMILES']
        enam_mol = Chem.MolFromSmiles(enam_smiles)
        enam_fp = AllChem.GetMorganFingerprintAsBitVect(enam_mol, 2)
        similarity = DataStructs.TanimotoSimilarity(ref_fp, enam_fp)

        # 결과 리스트에 추가
        results.append((index, ref_smiles, enam_row['CatalogID'], enam_smiles, similarity))

# 결과를 DataFrame으로 변환하고 유사도 순으로 정렬
result_df = pd.DataFrame(results, columns=['Ref_Index', 'Ref_SMILES', 'Enam_CatalogID', 'Enam_SMILES', 'Similarity'])
result_df = result_df.sort_values(by='Similarity', ascending=False)

# 결과 출력 및 저장
print(result_df)
result_df.to_csv("sorted_similarity_results.csv")
```

```python
import pandas as pd
from rdkit import Chem
from rdkit.Chem import AllChem
from rdkit.Chem import DataStructs

# 데이터 프레임 읽기
data_ref = pd.read_csv('/home/soowon/project/07_PPAR/ref_setA.csv')
data_enam = pd.read_csv('/home/soowon/Database/Enamine/SCR_2310/screen_lib2310.csv')


# data_ref의 모든 SMILES에 대해 반복
for index, ref_row in data_ref.iterrows():
    ref_smiles = ref_row['smiles']
    ref_mol = Chem.MolFromSmiles(ref_smiles)
    ref_fp = AllChem.GetMorganFingerprintAsBitVect(ref_mol, 2)

    # 결과를 저장할 리스트
    results = []

    # data_enam의 모든 SMILES에 대해 반복
    for enam_index, enam_row in data_enam.iterrows():
        enam_smiles = enam_row['SMILES']
        enam_mol = Chem.MolFromSmiles(enam_smiles)
        enam_fp = AllChem.GetMorganFingerprintAsBitVect(enam_mol, 2)
        similarity = DataStructs.TanimotoSimilarity(ref_fp, enam_fp)

        # 결과 리스트에 추가
        results.append((enam_row['CatalogID'], enam_smiles, similarity))

    # 결과를 DataFrame으로 변환하고 유사도 순으로 정렬
    result_df = pd.DataFrame(results, columns=['Enam_CatalogID', 'Enam_SMILES', 'Similarity'])
    result_df = result_df.sort_values(by='Similarity', ascending=False)

    # 각 리간드별로 결과 파일 저장
    file_name = f"similarity_results_for_{ref_row['name']}.csv"
    result_df.to_csv(file_name, index=False)

    print(f"Saved results for {ref_row['name']} to {file_name}")
```

```python
# 데이터 프레임 읽기
data_sim1 = pd.read_csv('similarity_results_for_SR11023.csv')
data_sim2 = pd.read_csv('similarity_results_for_DS45500853.csv')
```

```python
import pandas as pd

# 두 데이터 프레임 정렬
data_sim1.sort_values(by='Enam_CatalogID', inplace=True)
data_sim2.sort_values(by='Enam_CatalogID', inplace=True)

# 열 이름 변경
data_sim1.rename(columns={'Similarity': 'Similarity_SR11023'}, inplace=True)
data_sim2.rename(columns={'Similarity': 'Similarity_DS45500853'}, inplace=True)

# 데이터 합치기
combined_data = pd.merge(data_sim1, data_sim2[['Enam_CatalogID', 'Similarity_DS45500853']], on='Enam_CatalogID')

# Similarity 합 계산
combined_data['Total_Similarity'] = combined_data['Similarity_SR11023'] + combined_data['Similarity_DS45500853']

# Total_Similarity 기준으로 정렬
combined_data.sort_values(by='Total_Similarity', ascending=False, inplace=True)

# 결과 저장
combined_data.to_csv("combined_similarity_data.csv", index=False)

# 출력
print(combined_data)
```
